```python
plt.figure(figsize = (15, 8))

plt.subplot(2, 3, 1)
plt.hist(df['CRIM'], bins = 30, cumulative = False, color = 'k');
plt.title('Crime')

plt.subplot(2, 3, 2)
plt.hist(df['TAX'], bins = 30, cumulative = False, color = 'k');
plt.title('Tax')

plt.subplot(2, 3, 3)
plt.hist(df['PT'], bins= 30, cumulative = False, color = 'k');
plt.title('Pupil Teacher ratio')

plt.subplot(2, 3, 4)
plt.hist(df['CRIM'], bins = 30, cumulative = True, color = 'r');
plt.title('Crime')

plt.subplot(2, 3, 5)
plt.hist(df['TAX'], bins = 30, cumulative = True, color = 'r');
plt.title('Tax')

plt.subplot(2, 3, 6)
plt.hist(df['PT'], bins = 30, cumulative = True, color = 'r');
plt.title('Pupil Teacher ratio')

plt.tight_layout()
plt.show()
```
