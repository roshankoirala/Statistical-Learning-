### Continue exploring the data, and provide a brief summary of what you discover.

```python
sns.pairplot(df, vars = ['Apps', 'Accept', 'Enroll', 'Grad.Rate'])
plt.show()
```

```python
df['Accept_Rate'] = df['Accept'] / df['Apps']
df['Enroll_Rate'] = df['Enroll'] / df['Accept']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
plt.figure(figsize= (8, 6))

plt.subplot(2, 2, 1)
sns.scatterplot(x = 'Accept_Rate', y = 'Grad.Rate', data = df, color = 'r');
model.fit(df[['Accept_Rate']], y = df['Grad.Rate'])
plt.plot(df['Accept_Rate'], model.predict(df[['Accept_Rate']]), color = 'k')

plt.subplot(2, 2, 2)
sns.scatterplot('Accept_Rate', 'Enroll_Rate', data = df, color = 'r');
model.fit(df[['Accept_Rate']], y = df['Enroll_Rate'])
plt.plot(df['Accept_Rate'], model.predict(df[['Accept_Rate']]), color = 'k')

plt.subplot(2, 2, 3)
sns.scatterplot('Enroll_Rate', 'Grad.Rate', data = df, color = 'r');
model.fit(df[['Enroll_Rate']], y = df['Grad.Rate'])
plt.plot(df['Enroll_Rate'], model.predict(df[['Enroll_Rate']]), color = 'k')

plt.subplot(2, 2, 4)
sns.scatterplot('Apps', 'Grad.Rate', data = df, color = 'r');
model.fit(df[['Apps']], y = df['Grad.Rate'])
plt.plot(df['Apps'], model.predict(df[['Apps']]), color = 'k')

plt.tight_layout()
plt.show()
```

Acceptance rate and Graduation rate are negatively correlated. 

Acceptance rate and Enroll rate are positively correlated.

Enroll rate and graduate rate are positively correlated.

Application rate and graduation rate are positively correlated.
