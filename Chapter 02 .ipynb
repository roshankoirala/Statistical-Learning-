```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Ex 8

```python
college = pd.read_csv('data/College.csv')
college.head()
```

```python
college.columns
```

```python
college['Private']
```

```python
college.describe()
```

```python
sns.pairplot(college.iloc[:, 1:5])
plt.show()
```

```python
sns.boxplot(x='Private', y='Outstate', data=college)
plt.show()
```

```python
college['Elite'] = college['Top10perc'] > 50
college['Elite'].value_counts()
```

```python
sns.boxplot(x='Elite', y='Outstate', data=college)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
plt.subplots_adjust(wspace=0.3, hspace=0.3)

sns.histplot(x='Apps', data=college, bins=15, ax=axes[0, 0])
sns.histplot(x='Accept', data=college, bins=15, ax=axes[0, 1])
sns.histplot(x='Enroll', data=college, bins=15, ax=axes[1, 0])
sns.histplot(x='P.Undergrad', data=college, bins=15, ax=axes[1, 1])
plt.show()
```
